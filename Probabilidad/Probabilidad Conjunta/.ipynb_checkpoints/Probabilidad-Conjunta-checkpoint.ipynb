## Section 1: Introduccin

En este punto ya deberas tener una definicin de _"probabilidad"_ bastante completa que involucre todos los aspectos ms importantes. Sin embargo, para no quedarnos atascados antes de empezar, utilizaremos una definicin sencilla e iremos perfeccionando ms adelante: 
            
<div class="alert alert-success" style="color:black">
Tip: Una probabilidad es una fraccin de un conjunto finito.</div>

$$ \text{Probabilidad} = \frac{\text{nmero de resultados deseados}}{\text{nmero total de resultados posibles}} $$

<a id="s11"></a> 
### Step 1

Por ejemplo, si encuestamos a 1.000 personas y 20 de ellas son cajeros de banco, la fraccin que trabaja como cajero es del 0,02 o 2%. Si elegimos al azar a una persona de esta poblacin, la probabilidad de que sea cajero es del 2%. 
                
<div class="alert alert-success" style="color:black">
Tip: Con "al azar" quiero decir que todas las personas del conjunto de datos tienen la misma probabilidad de ser elegidas.</div>

$$ P(\text{banquero}) = \frac{20}{1000} = 0.02 = 2\% $$

<a id="s12"></a> 
### Step 2

Con esta pequea intro, pasemos al dataset para aplicarlo.
                
Usaremos *Pandas* para leer los datos y almacenarlos en un `DataFrame`.

```python
import pandas as pd

gss = pd.read_csv('data/gss_bayes.csv', index_col=0)
gss.head()
```

El `DataFrame` tiene una fila por cada persona encuestada y una columna por cada variable que he seleccionado.

<a id="s13"></a> 
### Step 3

Las columnas son:

* `caseid`: Id del encuestado (que es el ndice de la tabla).

* `year`: Ao en que se encuest a la persona.

* `age`: Edad de la persona cuando fue encuestada.

* `sex`: Masculino o femenino.

* `polviews`: Opiniones polticas en un rango de liberal a conservador.

* `partyid`: Afiliacin a un partido poltico, demcrata, independiente o republicano.

* `indus10`: Cdigo del sector en el que trabaja el encuestado.

Veamos estas variables con ms detalle, empezando por `indus10`.

## Section 2: Fraccin de banqueros

Ya cargado y analizado los datos en un `Dataframe`, procedamos a calcular la fraccin del total que sean cajeros de banco. 

<a id="s21"></a> 
### Step 1

El cdigo de "Banca y actividades relacionadas" es 6870, por lo que podemos seleccionar banqueros as:

```python
banquero = (gss['indus10'] == 6870)
banquero.head()
```

El resultado es una `Serie` de *Pandas* que contiene los valores booleanos `True` y `False`.

<a id="s22"></a> 
### Step 2

Si usamos la funcin `sum` en esta `Serie`, trata `True` como 1 y `False` como 0, por lo que el total es el nmero de banqueros.

```python
banquero.sum()
```

En este conjunto de datos hay 728 banqueros.

<a id="s23"></a> 
### Step 3

Para calcular la *fraccin* de banqueros, podemos utilizar la funcin `mean`, que calcula la fraccin de valores `True` en la `Serie`:

```python
banquero.mean()
```

Aproximadamente el 1,5% de los encuestados trabaja en la banca, por lo que si elegimos a una persona al azar del conjunto de datos, la probabilidad de que sea banquero es de aproximadamente el 1,5%.

## Section 3: La funcin de probabilidad

Cmo habrs visto en la seccin de Programacin, dividir el cdgo en **funciones** nos facilita la gestin de proyectos, adems de escribir software de manera eficaz.

<a id="s31"></a> 
### Step 1

Pondremos el cdigo de la seccin anterior en una funcin que toma una `Serie` booleana y devuelve una probabilidad:

```python
def prob(A):
    """Calcula la probabilidad de una proposicin, A."""
    return A.mean()
```

<a id="s32"></a> 
### Step 2

De esta manera, podemos calcular la fraccin de banqueros as:

```python
prob(banquero)
```

<a id="s33"></a> 
### Step 3

Ahora veamos otra variable en este conjunto de datos. Los valores de la columna `sex` estn codificados as:

    1    Male
    2    Female

<a id="s34"></a> 
### Step 4

As que podemos hacer una `Serie` booleana que es `True` para las mujeres encuestadas y `False` en caso contrario:

```python
femenino = (gss['sex'] == 2)
```

<a id="s35"></a> 
### Step 5

Y utilzala para calcular la fraccin de encuestados que son mujeres:

```python
prob(femenino)
```

<div class="alert alert-warning" style="color:black">
Nota: La fraccin de mujeres en este conjunto de datos es mayor que en la poblacin adulta de EE.UU. porque la GSS no incluye a las personas que viven en instituciones como prisiones y residencias militares, y esas poblaciones tienen ms probabilidades de ser masculinas.</div>

## Section 4: Partidos y opiniones polticas

En la seccin anterior vimos cmo calcular la probabilidad en la columna `age` que tiene slo dos valores posibles. A continuacin, usaremos dos columnas con ms valores posibles.

<a id="s41"></a> 
### Step 1

Una de las variables que podemos considerar es `polviews`, que describe las opiniones polticas de los encuestados.

Los valores de `polviews` estn en una escala de siete puntos:

    1	Extremely liberal (Extremadamente liberal)
    2	Liberal
    3	Slightly liberal (Ligeramente liberal)
    4	Moderate (Moderado)
    5	Slightly conservative (Ligeramente conservador)
    6	Conservative (Conservador)
    7	Extremely conservative (Extremadamente conservador)

<a id="s42"></a> 
### Step 2

Definir `liberal` como `True` para cualquiera cuya respuesta sea _Extremadamente liberal_ (`Extremely liberal`), _Liberal_ (`Liberal`) o _Ligeramente liberal_ (`Slightly liberal`):

```python
liberal = (gss['polviews'] <= 3)
```

<a id="s43"></a> 
### Step 3

Esta es la fraccin de encuestados que son liberales segn esta definicin:

```python
prob(liberal)
```

Si elegimos una persona al azar en este conjunto de datos, la probabilidad de que sea liberal es de aproximadamente el 27%.

<a id="s44"></a> 
### Step 4

La otra variable que vamos a considerar es `partyid`.

Los valores de `partyid` se codifican as:

    0	Strong democrat (Muy demcrata)
    1	Not strong democrat (No muy demcrata)
    2	Independent, near democrat (Independiente, casi demcrata)
    3	Independent (Independiente)
    4	Independent, near republican (Independiente, casi republicano)
    5	Not strong republican (No muy republicano)
    6	Strong republican (Republicano fuerte)
    7	Other party (Otro partido)

<a id="s45"></a> 
### Step 5

Definir demcrata para incluir a los encuestados que eligieron _Demcrata fuerte_ (`Strong democrat`) o _No demcrata fuerte_ (`Not strong democrat`):

```python
democrata = (gss['partyid'] <= 1)
```

<a id="s46"></a> 
### Step 6

Y aqu est la fraccin de encuestados que son demcratas, segn esta definicin:

```python
prob(democrata)
```

## Section 5: Probabilidad Conjunta

Ahora que tenemos una definicin de probabilidad y una funcin que la calcula, pasemos a la conjuncin.

_Conjuncin_ es otro nombre para la operacin lgica _y_ o _and_ en ingls. Si tenemos dos proposiciones, _A_ y _B_, la conjuncin _A_ y _B_ es _Verdadera_ si tanto _A_ como _B_ son _Verdaderas_, y _Falsa_ en caso contrario.

$$ P(A\text{ y } B) = P(A\cap B) = P(A,B) $$

<a id="s51"></a> 
### Step 1

Si tenemos dos `Series` booleanas, podemos utilizar el operador `&` para calcular su conjuncin. Por ejemplo, ya hemos calculado la probabilidad de que un encuestado sea _banquero_:

```python
prob(banquero)
```

<a id="s52"></a> 
### Step 2

Y la probabilidad de que sean demcratas:

```python
prob(democrata)
```

<a id="s53"></a> 
### Step 3

Ahora podemos calcular la probabilidad de que un encuestado sea banquero _**y**_ demcrata:

```python
prob(banquero & democrata)
```

Como era de esperar, la probabilidad (banquero y demcrata) es menor que la probabilidad (banquero), porque no todos los banqueros son demcratas.

<a id="s54"></a> 
### Step 4

Esperamos que la conjuncin sea **conmutativa**, es decir, que _A_ y _B_ sean lo mismo que _B_ y _A_. Para comprobarlo, tambin podemos calcular la probabilidad (demcrata y banquero):

```python
prob(democrata & banquero)
```

Como era de esperar, son iguales.
