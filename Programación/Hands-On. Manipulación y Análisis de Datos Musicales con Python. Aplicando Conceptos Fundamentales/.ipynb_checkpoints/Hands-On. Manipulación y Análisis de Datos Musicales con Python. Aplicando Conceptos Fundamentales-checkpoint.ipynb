### Step 2: Generacin del Conjunto de Datos

En este segundo paso, profundizaremos en la *historia detrs de los datos*. Exploraremos cmo se gener el conjunto de datos musical simulado que estamos a punto de analizar. Descubriremos el contexto detrs de estos datos ficticios y cmo se relacionan con el mundo de la msica. Este paso nos ayudar a comprender la naturaleza de los datos y por qu son tan interesantes de analizar.

<div class="alert alert-success" style="color:black">

**Tips:** 

En el mundo del anlisis de datos, crear conjuntos de **datos sintticos** es una prctica esencial. Nos permite desarrollar y probar algoritmos, modelos y tcnicas de anlisis de datos en un entorno controlado antes de aplicarlos a datos del mundo real. En este paso, descubriremos cmo se gener este conjunto de datos ficticios y cmo se relaciona con la prctica habitual en el anlisis y ciencia de datos. Comprenderemos el contexto detrs de estos datos sintticos y por qu son una herramienta valiosa en nuestro arsenal de anlisis de datos.

</div>

```python
# Crear un conjunto de datos vaco
data = []

# Definir el nmero de filas que deseas (por ejemplo, 200)
num_filas = 200

# Definir listas de valores para cada columna
titulos = ["Cancin " + str(i) for i in range(1, num_filas + 1)]
artistas = ["Artista " + str(i) for i in range(1, num_filas + 1)]
generos = ["Pop", "Rock", "Hip-Hop", "Electrnica", "Jazz"] * (num_filas // 5)

# Duracin aleatoria entre 120 y 420 segundos
duracion = [random.randint(120, 420) for _ in range(num_filas)]

# Popularidad aleatoria entre 50 y 100
popularidad = [random.randint(50, 100) for _ in range(num_filas)]

# Ao de lanzamiento aleatorio entre 1960 y 2022
lanzamiento = [random.randint(1960, 2022) for _ in range(num_filas)]

# Instrumentos aleatorios de una lista predefinida
instrumentos = ["Guitarra", "Batera", "Sintetizador", "Bajo", "Voz", "Piano", "Saxofn", "Percusin"] * (num_filas // 8)

# Construir el conjunto de datos fila por fila
for i in range(num_filas):
    fila = [titulos[i], artistas[i], generos[i], duracion[i], popularidad[i], lanzamiento[i], instrumentos[i]]
    data.append(fila)

# Crear un DataFrame de pandas
df = pd.DataFrame(data, columns=["Ttulo de la Cancin", "Artista", "Gnero", "Duracin (segundos)", "Popularidad", "Ao de Lanzamiento", "Instrumentos Utilizados"])

# Guardar el DataFrame como un archivo CSV
df.to_csv("conjunto_de_datos_musica_generado.csv", index=False)
```

### Step 2: Transformacin de datos

En este paso, nos adentraremos en la transformacin de datos. Aplicaremos funciones y estructuras de control para crear nuevas columnas derivadas o modificar las existentes. Por ejemplo, podemos convertir la duracin de las canciones de segundos a minutos o categorizar los aos de lanzamiento en dcadas. Estas transformaciones nos ayudarn a obtener informacin ms significativa y preparar los datos para el anlisis estadstico y la visualizacin.

```python
# Ejemplo de cdigo para convertir la duracin de segundos a minutos
df["Duracin (minutos)"] = df["Duracin (segundos)"] / 60

# Ejemplo de cdigo para categorizar los aos de lanzamiento en dcadas
def categorizar_decada(anio):
    if 1960 <= anio < 1970:
        return "1960s"
    elif 1970 <= anio < 1980:
        return "1970s"
    # ... continuar con ms dcadas

df["Dcada de Lanzamiento"] = df["Ao de Lanzamiento"].apply(categorizar_decada)
df.head()
```

### Step 1: Aplicacin de anlisis estadstico

En este primer paso, utilizaremos Python para realizar anlisis estadsticos bsicos en nuestros datos musicales. Calcularemos estadsticas descriptivas, como la media, la mediana y la desviacin estndar, para comprender mejor las caractersticas de las canciones en nuestro conjunto de datos. Estos anlisis nos proporcionarn informacin crucial sobre la distribucin de las duraciones, la popularidad y otros aspectos de las canciones.

1. Calculamos la duracin promedio de las canciones en segundos:

```python
duracion_promedio = df['Duracin (segundos)'].mean()
print("Duracin promedio de las canciones:", duracion_promedio, "segundos")
```

2. Calculando la popularidad mxima y mnima:

```python
# Encontramos la cancin ms y menos popular
cancion_mas_popular = df[df['Popularidad'] == df['Popularidad'].max()]
print("Cancin ms popular:")
cancion_mas_popular[['Ttulo de la Cancin', 'Artista', 'Popularidad']]
```

```python
cancion_menos_popular = df[df['Popularidad'] == df['Popularidad'].min()]
print("\nCancin menos popular:")
cancion_menos_popular[['Ttulo de la Cancin', 'Artista', 'Popularidad']]
```

3. Calculando la mediana de la popularidad por gnero musical:

```python
# Calculamos la mediana de la popularidad para cada gnero musical
mediana_popularidad_por_genero = df.groupby('Gnero')['Popularidad'].median()
print("Mediana de popularidad por gnero musical:")
mediana_popularidad_por_genero
```

### Step 2: Introduccin a la Visualizacin de datos

En el segundo paso, nos adentraremos en el mundo de la visualizacin de datos en Python. Aprenderemos cmo crear grficos y visualizaciones efectivas utilizando bibliotecas como Matplotlib y Seaborn. Exploraremos diferentes tipos de grficos, como histogramas, diagramas de dispersin y diagramas de barras, para representar y comunicar los resultados de nuestros anlisis de una manera clara y visualmente atractiva.

1: **Histograma de Duracin de Canciones**

Una de las formas ms simples y efectivas de visualizar datos numricos es a travs de un histograma. Aqu, crearemos un histograma para la columna de "Duracin (segundos)" de nuestro conjunto de datos musical. Esto nos permitir comprender mejor la distribucin de las duraciones de las canciones:

```python
# Crear un histograma de la duracin de las canciones
plt.hist(df['Duracin (segundos)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histograma de Duracin de Canciones')
plt.xlabel('Duracin (segundos)')
plt.ylabel('Nmero de Canciones')
plt.show()
```

2. **Grfico de Barras para Gneros Musicales**

Otra forma de visualizar datos es a travs de un grfico de barras. Aqu, crearemos un grfico de barras para mostrar la cantidad de canciones en cada gnero musical en nuestro conjunto de datos.

```python
# Contar la cantidad de canciones por gnero
conteo_generos = df['Gnero'].value_counts()

# Crear un grfico de barras
plt.figure(figsize=(10, 6))
conteo_generos.plot(kind='bar', color='salmon')
plt.title('Cantidad de Canciones por Gnero Musical')
plt.xlabel('Gnero Musical')
plt.ylabel('Nmero de Canciones')
plt.xticks(rotation=45)
plt.show()
```

3. **Grfico de Dispersin entre Duracin y Popularidad**:

Un grfico de dispersin nos permite visualizar la relacin entre dos variables numricas. Aqu, crearemos un grfico de dispersin para explorar si existe alguna relacin entre la duracin de las canciones y su popularidad.

```python
# Crear un grfico de dispersin entre duracin y popularidad
plt.figure(figsize=(8, 6))
plt.scatter(df['Duracin (segundos)'], df['Popularidad'], alpha=0.5, color='green')
plt.title('Grfico de Dispersin entre Duracin y Popularidad de Canciones')
plt.xlabel('Duracin (segundos)')
plt.ylabel('Popularidad')
plt.show()
```
