## Funciones - Intro

En este tutorial veremos cmo escribir funciones, que son bloques de cdigo con nombre, diseados para hacer un trabajo especfico. Cuando queremos realizar una tarea en particular que hemos definido en una funcin, llamamos a la funcin responsable de ello. 

Si necesitamos realizar esa tarea varias veces a lo largo de un programa, no necesitamos escribir todo el cdigo para la misma tarea una y otra vez; solo llamamos a la funcin dedicada de manejar esa tarea, y la llamada le dice a Python que ejecute el cdigo dentro de la funcin. 

El uso de funciones hace que nuestros  programas sean ms fciles de escribir, leer, probar y corregir.

Las funciones nos permiten escribir cdigo una vez y luego reutilizar ese cdigo tantas veces como queramos. Cuando necesitamos ejecutar el cdigo en una funcin, basta con escribir una llamada de una lnea y la funcin hace su trabajo.

Cuando necesitamos modificar el comportamiento de una funcin, solo tenemos que modificar el bloque de cdigo de su definicin, y ese cambio se refleja en todos los lugares donde llamemos a esa funcin.
El uso de funciones hace que nuestros programas sean ms fciles de leer, los buenos nombres de funciones resumen lo que hace cada parte de un programa. 

Leyendo una serie de llamadas a funciones tenemos idea mucho ms clara de qu hace un programa que leyendo una serie de bloques de cdigo.

### Step 3: Argumentos Posicionales

Cuando llamamos a una funcin, Python debe asociar cada argumento en la llamada con un parmetro en la definicin de la funcin. La forma ms simple de hacer esto es basado en el orden de los argumentos proporcionados. Los valores asociados de este modo a cada parmetro se llaman argumentos posicionales.

```python
# Definicin de la funcin 'describir_mascota'
def describir_mascota(tipo_animal, nombre_mascota):
    """Muestra informacin acerca de una mascota."""
    # Imprime el tipo de animal
    print(f"\nTengo un {tipo_animal}.")
    # Imprime el nombre de la mascota con la primera letra en mayscula
    print(f"El nombre de mi {tipo_animal} es {nombre_mascota.title()}.")

# Llamada a la funcin 'describir_mascota' con 'hamster' y 'harry' como argumentos
describir_mascota('hamster', 'harry')
# Llamada a la funcin 'describir_mascota' con 'perro' y 'willie' como argumentos
describir_mascota('perro', 'willie')
```

<a id="keyword"></a> 
### Step 4: Argumentos de Palabras Clave (Argumentos Keyword)

Un argumento keyword es un par nombre y valor que se pasa a una funcin. Asociamos el nombre y el valor dentro del argumento.
Los argumentos keyword nos liberan de tener que ordenar correctamente los argumentos en la llamada a la funcin, y especifican el papel de cada valor en la llamada a la funcin.

```python
# Llamada a la funcin 'describir_mascota' usando argumentos de palabras clave
# Aqu, se especifica explcitamente cada argumento con su nombre y valor
describir_mascota(tipo_animal='hamster', nombre_mascota='harry')

# Llamada a la misma funcin con los argumentos de palabras clave en orden diferente
# Los argumentos de palabras clave permiten cambiar el orden sin afectar la funcin
describir_mascota(nombre_mascota='harry', tipo_animal='hamster')
```

*Nota:
En este ejemplo, tipo_animal y nombre_mascota son argumentos de palabras clave. La principal ventaja de utilizar argumentos de palabras clave es que el orden en que se pasan los argumentos a la funcin ya no es importante, ya que cada argumento se especifica claramente por su nombre. Esto hace que el cdigo sea ms claro y fcil de entender, especialmente en funciones con muchos parmetros o cuando algunos de los valores pasados pueden no ser inmediatamente evidentes en su propsito.*

<a id="default"></a> 
### Step 5: Valores Predeterminados (Valores Default)

Al escribir una funcin, podemos definir un valor predeterminado para cada parmetro. Si se proporciona un argumento para un parmetro en la llamada a la funcin, Python usa ese valor del argumento. De lo contrario, utiliza el valor predeterminado del parmetro. Usar valores predeterminados puede simplificar las llamadas a funciones y aclarar las formas en que estas funcionese usan tpicamente.

```python
# Definicin de la funcin 'describir_mascota'
def describir_mascota(nombre_mascota, tipo_animal='perro'):
    """Muestra informacin acerca de una mascota."""
    # Imprime el tipo de animal. Usa el valor predeterminado si no se especifica.
    print(f"\nTengo un {tipo_animal}.")
    # Imprime el nombre de la mascota con la primera letra en mayscula.
    print(f"El nombre de mi {tipo_animal} es {nombre_mascota.title()}.")

# Llamada a la funcin 'describir_mascota' solo con el nombre de la mascota.
# Aqu se usa el valor predeterminado para el tipo de animal ('perro').
describir_mascota(nombre_mascota='willie')

# Llamada a la funcin con un solo argumento posicional.
# Python asume que este argumento corresponde al primer parmetro definido en la funcin.
describir_mascota('willie')
```

*Nota: En esta funcin, tipo_animal tiene un valor predeterminado de 'perro'. Esto significa que si no se especifica el tipo de animal al llamar a la funcin, se utilizar 'perro' por defecto. Los valores predeterminados permiten que las funciones sean ms flexibles y reducen la necesidad de especificar cada argumento en cada llamada a la funcin, simplificando as su uso.

<a id="llamadas"></a> 
### Step 6: Formas Equivalentes de Llamadas a Funcin

Este cdigo demuestra varias formas equivalentes de llamar a la misma funcin `describir_mascota`. Dependiendo de si se usan argumentos posicionales o de palabras clave (y en qu orden), estas llamadas a la funcin pueden ser flexibles pero seguirn siendo efectivas y producirn el mismo resultado. Esto ilustra cmo Python permite mltiples maneras de suministrar argumentos a las funciones, lo que proporciona una gran flexibilidad en la escritura de cdigo

```python
# Un perro llamado Willie.
# Llamada a la funcin usando solo un argumento posicional.
describir_mascota('willie')
# Llamada a la funcin especificando el nombre de la mascota con un argumento de palabra clave.
describir_mascota(nombre_mascota='willie')

# Un hmster llamado Harry.
# Llamada a la funcin con dos argumentos posicionales.
describir_mascota('harry', 'hmster')
# Llamada a la funcin especificando ambos argumentos con palabras clave.
describir_mascota(nombre_mascota='harry', tipo_animal='hmster')
# Llamada a la funcin con argumentos de palabras clave en orden invertido.
describir_mascota(tipo_animal='hmster', nombre_mascota='harry')
```

<a id="opcional"></a> 
### Step 7: Argumento Opcional

A veces tiene sentido hacer un argumento opcional para que podamos elegir proporcionar informacin adicional solo si tiene sentido. Usamos los valores predeterminados para hacer que un argumento sea opcional. En la primera funcin ningn argumento es opcional. En la segunda, `segundo_nombre` es opcional.

```python
# Definicin de la funcin 'obtener_nombre_formateado'
def obtener_nombre_formateado(primer_nombre, segundo_nombre, apellido):
    """Devuelve un nombre completo, formateado de manera ordenada."""
    # Construye el nombre completo
    nombre_completo = f"{primer_nombre} {segundo_nombre} {apellido}"
    # Devuelve el nombre con un formato ordenado
    return nombre_completo.title()

# Uso de la funcin con tres argumentos
musico = obtener_nombre_formateado('john', 'lee', 'hooker')
print(musico)

# Nueva definicin de la funcin con 'segundo_nombre' como argumento opcional
def obtener_nombre_formateado(primer_nombre, apellido, segundo_nombre=''):
    """Devuelve un nombre completo, formateado de manera ordenada."""
    # Comprueba si se proporcion un segundo nombre
    if segundo_nombre:
        nombre_completo = f"{primer_nombre} {segundo_nombre} {apellido}"
    else:
        nombre_completo = f"{primer_nombre} {apellido}"
    # Devuelve el nombre con un formato ordenado
    return nombre_completo.title()

# Uso de la funcin sin proporcionar un segundo nombre
musico = obtener_nombre_formateado('jimi', 'hendrix')
print(musico)

# Uso de la funcin proporcionando un segundo nombre
musico = obtener_nombre_formateado('john', 'hooker', 'lee')
print(musico)
```

*Nota: En este cdigo, la funcin obtener_nombre_formateado se define inicialmente con tres argumentos requeridos. Luego, se redefine con el segundo_nombre como un argumento opcional, utilizando un valor predeterminado vaco (''). Esto permite llamar a la funcin con o sin un segundo nombre. La funcin ajusta el formato del nombre completo dependiendo de si el segundo nombre est presente o no.*

<a id="arbitrarios"></a> 
### Step 8: Nmero Arbitrario de Argumentos

A veces no sabremos de antemano cuntos argumentos debe aceptar una funcin. Python permite que una funcin reciba un nmero arbitrario de argumentos en la llamada.

Por ejemplo, consideremos una funcin que construye una pizza. Debe aceptar una cantidad de ingredientes, pero no podemos saber de antemano cuntos ingredientes querr una persona.
La funcin en el siguiente ejemplo tiene un parmetro, * ingredientes, pero este parmetro acepta tantos argumentos como pasemos:

```python
# Definicin de la funcin 'hacer_pizza'
def hacer_pizza(*ingredientes):
    """Imprime la lista de ingredientes que se han solicitado."""
    # Imprime los ingredientes proporcionados
    print(ingredientes)

# Llamada a la funcin 'hacer_pizza' con un solo ingrediente
hacer_pizza('pepperoni')

# Llamada a la funcin 'hacer_pizza' con mltiples ingredientes
hacer_pizza('championes', 'pimientos verdes', 'queso extra')
```

*Nota: En este cdigo, *ingredientes en la definicin de hacer_pizza permite que la funcin acepte cualquier nmero de argumentos posicionales. Los argumentos pasados a la funcin se tratan como una tupla dentro de la funcin, lo que facilita su manipulacin. As, la funcin puede ser llamada con diferentes cantidades de ingredientes, demostrando la flexibilidad de Python para manejar un nmero arbitrario de argumentos en una funcin*.

<a id="kwarbitrarios"></a> 
### Step 9:Uso de Argumentos Keyword Arbitrarios

Podemos escribir funciones que acepten tantos pares clave-valor como pasemos en la llamada.

```python
# Definicin de la funcin 'crear_perfil'
def crear_perfil(nombre, apellido, **info_usuario):
    """Construye un diccionario que contiene todo lo que sabemos sobre un usuario."""

    # Agrega el nombre y el apellido al diccionario de informacin del usuario
    info_usuario['nombre'] = nombre
    info_usuario['apellido'] = apellido
    # Devuelve el diccionario completo
    return info_usuario

# Creacin de un perfil de usuario usando la funcin 'crear_perfil'
# Se pasan el nombre, el apellido y argumentos keyword arbitrarios
perfil_usuario = crear_perfil('albert', 'einstein',
                              ubicacion='princeton',
                              campo='fsica')
# Imprime el perfil del usuario
print(perfil_usuario)
```

*Nota: En este cdigo, info_usuario en la definicin de crear_perfil permite que la funcin acepte un nmero arbitrario de argumentos keyword (clave-valor). Estos argumentos son almacenados en un diccionario dentro 
de la funcin. El perfil de usuario se construye agregando el nombre y el apellido proporcionados al diccionario info_usuario, y luego se retorna este diccionario. La funcin puede ser llamada con diferentes pares clave-valor para almacenar variada informacin sobre un usuario*.

## Section 2: Mdulos

Una ventaja de las funciones es la forma en que separan los bloques de cdigo en nuestro programa principal. Al usar nombres descriptivos para las funciones, el programa principal ser mucho ms fcil de seguir. Podemos ir un paso ms all almacenando nuestras funciones en un archivo separado llamado mdulo y luego importando ese mdulo a nuestro programa principal. 

<a id="modulos1"></a> 
### Step 1: Mdulos y Separacin de Cdigo

La instruccin import dice a Python que nos permita disponer, en el archivo de programa actualmente en ejecucin, del cdigo en un mdulo. Almacenar nuestras funciones en un archivo separado nos permite ocultar los detalles del cdigo de nuestro programa y centrarnos en su lgica. Tambin nos permite reutilizar funciones en programas diferentes. Cuando almacenamos las funciones en archivos separados, podemos compartir esos archivos con otros programadores sin tener que compartir todo nuestro programa. Saber cmo importar funciones tambin nos permite usar bibliotecas de funciones desarrolladas por otros programadores.

```python
# Supongamos que tenemos un mdulo llamado 'car_creator.py' y dentro de este mdulo,
# hay una funcin llamada 'mostrar_informacion'.

# Enfoque 1: Importar el mdulo completo
import car_creator

# Llamamos a la funcin utilizando el nombre del mdulo
auto = car_creator.make_car('Toyota', 'Corolla', color='Rojo', ao=2021)
print('Enfoque 1\n', auto, '\n')

# Enfoque 2: Importar una funcin especfica del mdulo
from car_creator import make_car
# Llamamos a la funcin directamente
auto = make_car('Ford', 'Mustang', color='Azul', ao=2020)
print('Enfoque 2\n', auto, '\n')

# Enfoque 3: Importar una funcin especfica con un alias
from car_creator import make_car as crear_auto
# Llamamos a la funcin utilizando su alias
auto = crear_auto('Honda', 'Civic', color='Negro', ao=2022)
print('Enfoque 3\n', auto, '\n')

# Enfoque 4: Importar el mdulo completo con un alias
import car_creator as cc
# Llamamos a la funcin utilizando el alias del mdulo
auto = cc.make_car('Chevrolet', 'Camaro', color='Amarillo', ao=2019)
print('Enfoque 4\n', auto, '\n')

# Enfoque 5: Importar todas las funciones del mdulo directamente
from car_creator import *
# Llamamos a la funcin directamente
auto = make_car('Tesla', 'Model S', color='Blanco', ao=2021)
print('Enfoque 5\n', auto, '\n')
```

<a id="modulos2"></a> 
### Step 2: Importar un Mdulo Completo

Para comenzar a importar funciones, primero necesitamos crear un mdulo. Un mdulo es un archivo que termina en .py que contiene el cdigo que desea importar en su programa. Al usar **`import pizza`**, importamos todas las funciones definidas en el archivo pizza.py. Para llamar a una funcin desde un mdulo importado, ingresamos el nombre del mdulo que importamos, pizza, seguido del nombre de la funcin, make_pizza(), separados por un punto: **`pizza.make_pizza()`**. Esta forma de import disponibiliza en nuestro programa todas las funciones del mdulo. Si usamos esta forma de importar un mdulo completo llamado module_name.py, cada funcin en el mdulo estar disponible a travs de la siguiente sintaxis: **`module_name.function_name()`**.

```python
# Importamos el mdulo completo 'pizza'
import pizza

# Llamamos a la funcin 'hacer_pizza' del mdulo 'pizza' con un solo ingrediente: pepperoni
pizza.hacer_pizza('pepperoni')

# Llamamos a la misma funcin con varios ingredientes: championes, pimientos verdes y queso extra
pizza.hacer_pizza('championes', 'pimientos verdes', 'queso extra')
```

<a id="modulos3"></a> 
### Step 3: Importar Funciones Especficas

Tambin podemos importar una funcin especfica desde un mdulo. La sintaxis para esto es: **`from module_name import function_name`**. Podemos importar tantas funciones como queramos de un mdulo, separando cada nombre de funcin por coma: **`from module_name import function_0, function_1, function_2`**.

```python
# Importamos la funcin 'hacer_pizza' del mdulo 'pizza'
from pizza import hacer_pizza

# Llamamos a la funcin 'hacer_pizza' con un solo ingrediente: pepperoni
hacer_pizza('pepperoni')

# Llamamos a la misma funcin con varios ingredientes: championes, pimientos verdes y queso extra
hacer_pizza('championes', 'pimientos verdes', 'queso extra')
```
